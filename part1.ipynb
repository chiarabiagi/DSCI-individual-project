# 3. Exploratory Data Analysis and Visualization
Do the minimum necessary wrangling to turn your data into a tidy format. Do not do any additional wrangling here; that will happen later during the group project phase.
Compute the mean value foreach quantitative variable in the players.csv data set. Report the mean values in a table format.

```r
library(tidyverse)

# Computing the mean for each quantitative variable in players.csv 
players |>
  select(Age, played_hours) |>
  map_dfr(mean, na.rm = TRUE)



players_small <- players |>
select(Age, gender, played_hours)|>
  drop_na(Age) |>    # remove rows with NA in the column Age
  filter(played_hours >1) 
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)

gender_plot <- ggplot(players_small, aes(x = gender, y = played_hours, fill = gender)) +
  geom_bar(stat = "identity") +
  xlab("Gender") +
  ylab("Number of Hours Played (hrs)") +
  labs(fill = "Gender", title = "Gender vs Number of Hours Played") +
  theme(text = element_text(size = 14))

gender_plot
```

### Insights I gain from the plot Gender vs Number of Hours Played

The goal of this plot is to know which gender plays the most amount of video games. From this plot, I see that the majority is male and female. Now that I know which genders to target, I want to know which age out of males and females play the most. 

```r
options(repr.plot.width = 8, repr.plot.height = 8)

age_plot <- players_small |>
  filter(gender %in% c("Male", "Female")) |>  
  ggplot(aes(x = Age, y = played_hours, colour = gender, shape = gender)) +
  geom_point(alpha = 0.8) +
  scale_y_log10() +
  scale_colour_brewer(palette = "Set1") + 
  xlab("Age of Players (yrs)") +
  ylab("Number of Hours Played (hrs)") +
  labs(colour = "Gender", shape = "Gender",
  title= "Age vs Number of Hours Played in Males and Females") +
  theme(text = element_text(size = 14))

age_plot
```

### Insights I gain from the plot age versus number of hours played in males and females

This plot shows that ages between around 17 and 24 for both males and females play the most.
